```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to load the image data for each frame
def load_image(frame_index, video_key):
    image_path = f"/nas/Dataset/Phoenix/phoenix2014-release/phoenix-2014-multisigner/features/{video_key}.avi_pid0_fn000{frame_index:03d}-0.png"
    return cv2.imread(image_path)

# Define the finger labels
finger_labels = {
    range(91, 96): 'T',   # Thumb points T1â€“T5
    range(96, 100): 'I',  # Index points I1â€“I4
    range(100, 104): 'M',  # Middle points M1â€“M4
    range(104, 108): 'R',  # Ring points R1â€“R4
    range(108, 112): 'P'   # Pinkie points P1â€“P4
}

# Iterate through each video
for video_key, video_data in data.items():
    keypoints = video_data['keypoints']

    # Only visualize for a specific video file
    if video_key == "fullFrame-210x260px/train/01April_2010_Thursday_heute_default-0/1/01April_2010_Thursday_heute":
        # Iterate through each frame, limit to first 10 frames
        for frame_index, frame_keypoints in enumerate(keypoints[:10]):  # Limit to first 10 frames
            frame_image = load_image(frame_index, video_key)  # Load the corresponding image
            frame_image = cv2.cvtColor(frame_image, cv2.COLOR_BGR2RGB)  # Convert to RGB

            # Separate the keypoints for the left hand (indices 91-111)
            left_hand_keypoints = frame_keypoints[91:112]

            # Create the plot
            plt.figure(figsize=(8, 6))

            # Plot left hand keypoints (in blue)
            plt.scatter(left_hand_keypoints[:, 0], -left_hand_keypoints[:, 1], color='blue', s=5, label='Left Hand')  # Invert y for correct visualization

            # Add custom labels based on finger segments
            for i, (x, y, c) in enumerate(left_hand_keypoints):
                # Determine the label prefix (T, I, M, R, P) based on the index
                for key_range, prefix in finger_labels.items():
                    if (91 + i) in key_range:
                        label = f"{prefix}{(91 + i) - min(key_range) + 1}"  # Determine the suffix (1, 2, etc.)
                        plt.text(x, -y, label, fontsize=8, color='red')  # Invert y for correct placement

            # Customize plot
            plt.title(f"Frame {frame_index} - Left Hand Keypoints")
            plt.xlabel('x-coordinate')
            plt.ylabel('y-coordinate')
            plt.legend()
            plt.gca().invert_yaxis()  # Invert y-axis for correct visual alignment
            plt.show()
        
        # Break out of the loop after visualizing the first 10 frames
        break
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to load the image data for each frame
def load_image(frame_index, video_key):
    image_path = f"/nas/Dataset/Phoenix/phoenix2014-release/phoenix-2014-multisigner/features/{video_key}.avi_pid0_fn000{frame_index:03d}-0.png"
    return cv2.imread(image_path)

# Define colors for each thumb keypoint trajectory
colors = ['blue', 'green', 'orange', 'purple', 'brown']  # T1 to T5

# Iterate through each video
for video_key, video_data in data.items():
    keypoints = video_data['keypoints']
    
    # Only visualize for a specific video file
    if video_key == "fullFrame-210x260px/train/01April_2010_Thursday_heute_default-0/1/01April_2010_Thursday_heute":
        
        # Visualize thumb keypoint trajectories for frames 0 and 1 in one window
        plt.figure(figsize=(8, 6))

        # Initialize lists to store each thumb keypoint's trajectory across frames
        thumb_trajectories_x = [[] for _ in range(5)]
        thumb_trajectories_y = [[] for _ in range(5)]

        for frame_index in range(2):  # First 2 frames
            frame_keypoints = keypoints[frame_index]
            thumb_keypoints = frame_keypoints[91:96]
            
            # Extract x and y for each thumb point (T1 to T5) and store in trajectory lists
            for i, (x, y, _) in enumerate(thumb_keypoints):
                thumb_trajectories_x[i].append(x)
                thumb_trajectories_y[i].append(-y)  # Invert y for natural orientation

            # Plot current frame's thumb keypoints as dots
            plt.scatter([kp[0] for kp in thumb_keypoints], [-kp[1] for kp in thumb_keypoints], 
                        color='black', s=10, label=f'Frame {frame_index}' if frame_index == 0 else "")

        # Draw dashed trajectory lines connecting the same thumb keypoints across frames
        for i in range(5):  # T1 to T5
            plt.plot(thumb_trajectories_x[i], thumb_trajectories_y[i], linestyle='--', color=colors[i], label=f'T{i+1}')
            plt.text(thumb_trajectories_x[i][0], thumb_trajectories_y[i][0], f'T{i+1}', fontsize=8, color=colors[i])

        plt.title("Thumb Keypoints Trajectory - Frames 0 and 1")
        plt.xlabel('x-coordinate')
        plt.ylabel('y-coordinate')
        plt.legend()
        plt.gca().invert_yaxis()
        plt.show()

        # Repeat for frames 0, 1, and 2 in a new window
        plt.figure(figsize=(8, 6))
        thumb_trajectories_x = [[] for _ in range(5)]
        thumb_trajectories_y = [[] for _ in range(5)]

        for frame_index in range(3):  # First 3 frames
            frame_keypoints = keypoints[frame_index]
            thumb_keypoints = frame_keypoints[91:96]

            for i, (x, y, _) in enumerate(thumb_keypoints):
                thumb_trajectories_x[i].append(x)
                thumb_trajectories_y[i].append(-y)

            plt.scatter([kp[0] for kp in thumb_keypoints], [-kp[1] for kp in thumb_keypoints], 
                        color='black', s=10, label=f'Frame {frame_index}' if frame_index == 0 else "")

        for i in range(5):
            plt.plot(thumb_trajectories_x[i], thumb_trajectories_y[i], linestyle='--', color=colors[i], label=f'T{i+1}')
            plt.text(thumb_trajectories_x[i][0], thumb_trajectories_y[i][0], f'T{i+1}', fontsize=8, color=colors[i])

        plt.title("Thumb Keypoints Trajectory - Frames 0 to 2")
        plt.xlabel('x-coordinate')
        plt.ylabel('y-coordinate')
        plt.legend()
        plt.gca().invert_yaxis()
        plt.show()

        # Stop after visualizing the required frames
        break
```
