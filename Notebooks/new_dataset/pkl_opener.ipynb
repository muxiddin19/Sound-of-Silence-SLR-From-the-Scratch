```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

def filter_frames_with_threshold(pkl_file_path, output_pkl_path, threshold_function):
    """
    Filter unnecessary frames from a pkl dataset based on a threshold function.

    Parameters:
        pkl_file_path (str): Path to the input pkl file.
        output_pkl_path (str): Path to save the filtered pkl file.
        threshold_function (function): Function to calculate and apply the threshold logic.
    """
    # Load the pkl file
    with open(pkl_file_path, 'rb') as f:
        data = pickle.load(f)
    
    # Assume 'pose' contains the keypoint data, specifically the right elbow (index 3, Y-coordinate)
    pose_data = data['pose']
    right_elbow_y = pose_data[:, 3, 1]  # Extract the Y-coordinates of the right elbow across all frames

    # Apply the threshold function to determine valid gesture frames
    gesture_frames = threshold_function(right_elbow_y)

    # Filter the dataset to include only gesture frames
    filtered_data = {key: value[gesture_frames] for key, value in data.items()}

    # Save the filtered dataset
    with open(output_pkl_path, 'wb') as f:
        pickle.dump(filtered_data, f)

    print(f"Filtered dataset saved to: {output_pkl_path}")

    # Visualize the gesture detection for validation
    plt.figure(figsize=(10, 6))
    plt.plot(right_elbow_y, label="Right Elbow Y-coordinate")
    plt.axhline(np.mean(right_elbow_y) - np.std(right_elbow_y), color="red", linestyle="--", label="Threshold")
    plt.scatter(gesture_frames, right_elbow_y[gesture_frames], color="green", label="Gesture Frames", zorder=5)
    plt.title("Gesture Detection Based on Threshold")
    plt.xlabel("Frame Number")
    plt.ylabel("Y-coordinate")
    plt.legend()
    plt.show()


# Example threshold function
def threshold_function(right_elbow_y):
    """
    Detect gesture start and end based on a threshold.

    Parameters:
        right_elbow_y (np.ndarray): Array of right elbow Y-coordinates.

    Returns:
        list: Indices of valid gesture frames.
    """
    mean_value = np.mean(right_elbow_y)
    std_dev = np.std(right_elbow_y)
    threshold = mean_value - std_dev  # Suggested threshold

    gesture_start = None
    gesture_end = None
    gesture_frames = []  # Store all valid gesture frames

    for i in range(len(right_elbow_y)):
        if right_elbow_y[i] <= threshold:  # Threshold satisfied
            if gesture_start is None:  # Start gesture
                gesture_start = i
            gesture_frames.append(i)  # Add frame to gesture
        elif gesture_start is not None:  # Gesture ongoing but current frame doesn't satisfy threshold
            # Check if any future frames satisfy the threshold
            future_frames_satisfy = any(
                right_elbow_y[j] <= threshold
                for j in range(i + 1, len(right_elbow_y))
            )
            if not future_frames_satisfy:  # If no future frames satisfy the threshold, stop the gesture
                gesture_end = i - 1
                break

    # Final gesture frame range
    if gesture_start is not None and gesture_end is None:
        gesture_end = len(right_elbow_y) - 1  # If gesture continues till the last frame

    return list(range(gesture_start, gesture_end + 1))


# Example usage
pkl_file_path = "/nas/Chingiz/sing_language/newDATA_points/restructured_by_parts.pkl"
output_pkl_path = "/nas/Chingiz/sing_language/newDATA_points/output_filtered_threshold.pkl"
filter_frames_with_threshold(pkl_file_path, output_pkl_path, threshold_function)
```
