```python
import zipfile
import json

# Path to the ZIP file
zip_file_path = '/nas/Dataset/KSL/Train/[라벨]01_crowd_keypoint.zip'

# Path inside the ZIP to the folder containing JSON files
folder_in_zip = '01_crowd_keypoint/01/NIA_SL_FS0001_CROWD01_F/'

# Desired file index (0-based)
target_index = 130  # Change this to the desired index

# Open the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as z:
    # List all JSON files in the folder
    json_files = [f for f in z.namelist() if f.startswith(folder_in_zip) and f.endswith('.json')]
    
    # Sort files numerically by frame number
    sorted_json_files = sorted(json_files, key=lambda x: int(x.split('_')[-2]))
    
    # Check if the index is within the range
    if 0 <= target_index < len(sorted_json_files):
        # Get the file at the desired index
        target_file = sorted_json_files[target_index]
        
        # Read the specific JSON file
        with z.open(target_file) as file:
            data = json.load(file)
            print(f"Data from file {target_index} ({target_file}): {data}")
    else:
        print(f"Index {target_index} is out of range. Total files: {len(sorted_json_files)}")
```
