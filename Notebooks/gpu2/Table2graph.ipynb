```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# Complete data from the table
papers = [
    'Aditya et al. [29]', 'Chen et al. [34]', 'Chen et al. [8]', 'Zuo et al. [9]',
    'Liu et al. [35]', 'Zuo et al. [16]', 'Guo et al. [23]', 'Jang et al. [30]',
    'Zheng et al. [17]', 'Hu et al. [20]', 'Hu et al. [21]', 'Xue et al. [19]',
    'Zuo et al. [11]', 'Zhu et al. [31]', 'Zhu et al. [46]', 'Gan et al. [37]'
]

publications = [
    'Sensors 2022', 'PRAI 2022', 'NeurIPS 2022', 'CVPR 2022',
    'HDIS 2022', 'ISCA 2022', 'CVPR 2023', 'ICASSP 2023',
    'CVPR 2023', 'CVPR 2023', 'AAAI 2023', 'M.T.A. 2023',
    'ACM 2024', 'S.E. 2023', 'E.I. 2024', 'CVPR 2024'
]

# Create empty dataframes for DEV and TEST
columns = ['Publication', 'Paper', 'CSL', 'CSL-Daily', 'PHOENIX2014', 'PHOENIX2014T']
dev_df = pd.DataFrame(columns=columns)
test_df = pd.DataFrame(columns=columns)

# Fill dataframes with data
dev_data = [
    [publications[0], papers[0], None, None, 20.5, 20.5],
    [publications[1], papers[1], None, None, 20.1, 20.1],
    [publications[2], papers[2], 25.4, 25.4, 18.4, 18.2],
    [publications[3], papers[3], None, None, 20.5, 20.2],
    [publications[4], papers[4], None, None, 25.9, 25.5],
    [publications[5], papers[5], None, None, 21.4, 21.1],
    [publications[6], papers[6], 29.0, 29.4, 18.6, 18.2],
    [publications[7], papers[7], 33.1, 33.1, 20.9, 20.4],
    [publications[8], papers[8], 30.6, 30.6, 19.4, 19.1],
    [publications[9], papers[9], 29.3, 29.3, 18.8, 18.6],
    [publications[10], papers[10], 30.7, 30.7, 19.5, 19.2],
    [publications[11], papers[11], 30.1, 30.1, 26.3, 26.0],
    [publications[12], papers[12], 31.0, 31.0, 20.5, 20.2],
    [publications[13], papers[13], 29.4, 29.4, 21.4, 21.1],
    [publications[14], papers[14], 27.8, 27.8, 20.3, 20.0],
    [publications[15], papers[15], 24.9, 24.9, 18.1, 18.1]
]

test_data = [
    [publications[0], papers[0], None, None, 21.5, 21.5],
    [publications[1], papers[1], None, None, 21.2, 21.2],
    [publications[2], papers[2], 25.3, 25.3, 18.8, 18.4],
    [publications[3], papers[3], None, None, 20.4, 20.2],
    [publications[4], papers[4], None, None, 26.4, 26.0],
    [publications[5], papers[5], None, None, 21.9, 21.5],
    [publications[6], papers[6], 29.1, 29.6, 19.0, 18.6],
    [publications[7], papers[7], 33.3, 33.3, 20.8, 20.4],
    [publications[8], papers[8], 30.8, 30.8, 20.3, 20.0],
    [publications[9], papers[9], 29.5, 29.5, 19.4, 19.2],
    [publications[10], papers[10], 31.0, 31.0, 21.0, 20.8],
    [publications[11], papers[11], 30.3, 30.3, 25.7, 25.4],
    [publications[12], papers[12], 31.2, 31.2, 20.4, 20.1],
    [publications[13], papers[13], 29.7, 29.7, 20.9, 20.5],
    [publications[14], papers[14], 27.9, 27.9, 21.4, 21.1],
    [publications[15], papers[15], 25.1, 25.1, 20.1, 19.8]
]

dev_df = pd.DataFrame(dev_data, columns=columns)
test_df = pd.DataFrame(test_data, columns=columns)

# Extract years from publications for coloring
years = [pub.split()[-1] for pub in publications]
unique_years = sorted(set(years))
year_colors = {'2022': '#FF5252', '2023': '#4CAF50', '2024': '#2196F3'}

# Set up bright color palette
colors = sns.color_palette("bright", 16)

# Plot bar charts with improved design
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
datasets = ['CSL', 'CSL-Daily', 'PHOENIX2014', 'PHOENIX2014T']

for i, dataset in enumerate(datasets):
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    
    # Gather data
    dev_values = []
    test_values = []
    labels = []
    paper_colors = []
    
    for j, paper in enumerate(papers):
        dev_val = dev_df.iloc[j][dataset]
        test_val = test_df.iloc[j][dataset]
        
        if pd.notna(dev_val) and pd.notna(test_val):
            dev_values.append(dev_val)
            test_values.append(test_val)
            paper_year = publications[j].split()[-1]
            paper_colors.append(year_colors[paper_year])
            labels.append(f"{paper.split()[0]} ({publications[j].split()[0]})")
    
    x = np.arange(len(labels))
    width = 0.35
    
    # Plot bars
    dev_bars = ax.bar(x - width/2, dev_values, width, label='DEV', color=paper_colors, edgecolor='black', linewidth=1, alpha=0.8)
    test_bars = ax.bar(x + width/2, test_values, width, label='TEST', color=paper_colors, edgecolor='black', linewidth=1, hatch='//', alpha=0.8)
    
    # Add data labels
    for bar in dev_bars:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.3,
                f'{height:.1f}', ha='center', va='bottom', fontsize=8, rotation=0, fontweight='bold')
    
    for bar in test_bars:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.3,
                f'{height:.1f}', ha='center', va='bottom', fontsize=8, rotation=0, fontweight='bold')
    
    # Customize appearance
    ax.set_title(f'{dataset} Dataset - WER (%)', fontsize=14, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=10)
    ax.set_ylim(0, max(max(dev_values, default=0), max(test_values, default=0)) * 1.2)
    ax.set_ylabel('Word Error Rate (%)', fontsize=12)
    ax.grid(axis='y', linestyle='--', alpha=0.3)
    ax.legend(fontsize=10)
    
    # Add best performance indicators
    if dev_values:
        min_dev_idx = dev_values.index(min(dev_values))
        min_test_idx = test_values.index(min(test_values))
        
        min_dev_bar = dev_bars[min_dev_idx]
        min_test_bar = test_bars[min_test_idx]
        
        ax.text(min_dev_bar.get_x() + min_dev_bar.get_width()/2., min_dev_bar.get_height() - 2.5,
                '★', ha='center', va='bottom', fontsize=14, color='red')
        ax.text(min_test_bar.get_x() + min_test_bar.get_width()/2., min_test_bar.get_height() - 2.5,
                '★', ha='center', va='bottom', fontsize=14, color='red')

# Add a year legend
labels = []
handles = []
for year, color in year_colors.items():
    labels.append(f'{year}')
    handles.append(plt.Rectangle((0,0), 1, 1, color=color))
    
fig.legend(handles, labels, loc='upper center', title='Publication Year', 
           bbox_to_anchor=(0.5, 0.98), ncol=3, fontsize=12, title_fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.suptitle('Word Error Rate (WER) Comparison Across Sign Language Recognition Models', 
             fontsize=16, fontweight='bold', y=0.99)

# Create heatmap for the performance table
fig2, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 16))

# Prepare heatmap data
datasets = ['CSL', 'CSL-Daily', 'PHOENIX2014', 'PHOENIX2014T']
combined_data = []

for i in range(len(papers)):
    row = []
    for dataset in datasets:
        row.append(dev_df.iloc[i][dataset])
    for dataset in datasets:
        row.append(test_df.iloc[i][dataset])
    combined_data.append(row)

combined_df = pd.DataFrame(combined_data, index=[f"{p} ({pub})" for p, pub in zip(papers, publications)], 
                          columns=['DEV-'+d for d in datasets] + ['TEST-'+d for d in datasets])

# Create a custom colormap - lower WER is better (green), higher WER is worse (red)
cmap = LinearSegmentedColormap.from_list('WER_cmap', ['#2ecc71', '#f1c40f', '#e74c3c'])

# Find the range for the colormap
vmin = combined_df.min().min()
vmax = combined_df.max().max()

# Create heatmap with improved design
sns.heatmap(combined_df, annot=True, cmap=cmap, linewidths=.5, ax=ax1, 
            vmin=vmin, vmax=vmax, fmt='.1f', cbar_kws={'label': 'WER (%)'})
ax1.set_title('DEV and TEST Word Error Rate (%) by Model and Dataset', fontsize=16, fontweight='bold')
ax1.set_ylabel('Model (Publication)', fontsize=12)

# Create a second heatmap showing the rank of each model for each dataset
rank_df = combined_df.copy()
for col in rank_df.columns:
    # Only rank non-NaN values
    valid_vals = rank_df[col].dropna()
    if not valid_vals.empty:
        ranks = valid_vals.rank()
        for idx in ranks.index:
            rank_df.loc[idx, col] = ranks[idx]
    
# Create rank heatmap
sns.heatmap(rank_df, annot=True, cmap='YlGnBu_r', linewidths=.5, ax=ax2, fmt='.0f', 
            cbar_kws={'label': 'Rank (1 = Best)'})
ax2.set_title('Ranking of Models by WER Performance (1 = Best)', fontsize=16, fontweight='bold')
ax2.set_ylabel('Model (Publication)', fontsize=12)

plt.tight_layout()
plt.show()
```
