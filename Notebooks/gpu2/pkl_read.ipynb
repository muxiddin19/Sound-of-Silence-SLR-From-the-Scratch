```python
import pickle

def read_pkl(path):
    with open(path, 'rb') as f:
        try:
            data1 = pickle.load(f, encoding='latin1')  # Try specifying encoding
        except Exception as e:
            print(f"Error loading pickle: {e}")
            raise
    return data1
```

```python
path = '/nas/Dataset/Phoenix/deltas.pkl'
data = read_pkl(path)
data
```
